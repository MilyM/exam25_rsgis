```python
import numpy as np
import matplotlib.pyplot as plt
import rasterio
```

```python
folder_path = r"E:\AAmagisterka\S2\EXAM_data"
ndvi1_path = f"{folder_path}\\t5_ndvi2024_Jul_Aug.tif"
ndvi2_path = f"{folder_path}\\t5_ndvi2024May.tif"
```

```python
def read_raster(file_path):
    with rasterio.open(file_path) as src:
        return src.read(1), src.profile
```

```python
ndvi1, profile1 = read_raster(ndvi1_path)
ndvi2, profile2 = read_raster(ndvi2_path)
ndvi_diff = ndvi1 - ndvi2
```

```python
def display_image(data, title, cmap='viridis'):
    plt.figure(figsize=(6, 5))
    plt.imshow(data, cmap=cmap)
    plt.colorbar()
    plt.title(title)
    plt.show()
```

```python
display_image(ndvi1, "NDVI - July/August 2024", cmap='Greens')
display_image(ndvi2, "NDVI - May 2024", cmap='Greens')
display_image(ndvi_diff, "NDVI Difference (July/August - May)", cmap='coolwarm')
```

```python
def plot_histogram(data, title):
    plt.figure(figsize=(6, 4))
    plt.hist(data.ravel(), bins=50, color='green', alpha=0.7)
    plt.title(title)
    plt.xlabel("NDVI Value")
    plt.ylabel("Frequency")
    plt.show()
```

```python
plot_histogram(ndvi1, "Histogram of NDVI - July/August 2024")
plot_histogram(ndvi2, "Histogram of NDVI - May 2024")

```

```python
def plot_scatter(x_data, y_data, title):
    plt.figure(figsize=(6, 5))
    plt.scatter(x_data.ravel(), y_data.ravel(), alpha=0.3, c='blue')
    plt.xlabel("NDVI - May 2024")
    plt.ylabel("NDVI - July/August 2024")
    plt.title(title)
    plt.show()
```

```python
plot_scatter(ndvi2, ndvi1, "Scatter Plot: NDVI May vs NDVI July/August")
```

```python
import os 
import rasterio
import matplotlib.pyplot as plt
```

```python
data_path = 'C:/egzamin/data/'

files = os.listdir(data_path)
files
```

```python
def load_img(path):
    with rasterio.open(path) as img:
        return img.read(1)
```

```python
imgs = []
for each in files:
    path = data_path + each
    # print(path)
    img = load_img(data_path + each)
    imgs.append(img)
```

```python
def display(img, colors, title):
    plt.imshow(img, cmap=colors)
    plt.title(title)
    plt.show()
```

```python
def histogram(data, title, bins=50):
    plt.hist(data.ravel(), bins=bins, color='green', alpha=0.9)
    plt.xlabel('Value')
    plt.title(title)
    plt.show()
```

```python
def scatter(x, y):
    plt.figure(figsize=(6, 5))
    plt.scatter(x.ravel(), y.ravel(), alpha=0.5, c='blue', s=1)
    plt.xlabel('NDVI')
    plt.ylabel('temp')
    plt.title("NDVI TEMP SCATTER")
    plt.show()
```

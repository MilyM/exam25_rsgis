```python
#Task1 Weronika Budzisz
#Importing data

import rasterio
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap


with rasterio.open(r"C:\egz_python\files1\t4_lst2023_Jul_May.tif") as lst_2023, \
     rasterio.open(r"C:\egz_python\files1\t4_ndvi_2024_Jul_Aug.tif") as ndvi_2023, \
     rasterio.open(r"C:\egz_python\files1\t4_lst2024May.tif") as lst_2024, \
     rasterio.open(r"C:\egz_python\files1\t4_ndvi_2024May.tif") as ndvi_2024:

    lst23 = lst_2023.read(1)
    ndvi23 = ndvi_2023.read(1)
    lst24 = lst_2024.read(1)
    ndvi24 = ndvi_2024.read(1)
```

```python
#Palett colors
def create_palette(colors):
    return LinearSegmentedColormap.from_list("custom_palette", colors, N=256)

temp_palette = create_palette(["blue", "cyan", "green", "yellow", "red"])  # Temperature
ndvi_palette = create_palette(["brown", "yellow", "green"])  # NDVI
diff_palette = create_palette(["black", "white", "pink"])  # Difference example
```

```python
# Display images for 2023
fig, ax = plt.subplots(1, 3, figsize=(18, 6))

im1 = ax[0].imshow(lst23, cmap=temp_palette)
ax[0].set_title("Temperature 2023")
plt.colorbar(im1, ax=ax[0])

im2 = ax[1].imshow(ndvi23, cmap=ndvi_palette)
ax[1].set_title("NDVI 2023")
plt.colorbar(im2, ax=ax[1])

diff23 = lst23 - ndvi23
im3 = ax[2].imshow(diff23, cmap=diff_palette)
ax[2].set_title("Difference (Temp - NDVI) 2023")
plt.colorbar(im3, ax=ax[2])

plt.tight_layout()
plt.show()

```

```python
# Display images for 2024
fig, ax = plt.subplots(1, 3, figsize=(18, 6))

im1 = ax[0].imshow(lst24, cmap=temp_palette)
ax[0].set_title("Temperature 2024")
plt.colorbar(im1, ax=ax[0])

im2 = ax[1].imshow(ndvi24, cmap=ndvi_palette)
ax[1].set_title("NDVI 2024")
plt.colorbar(im2, ax=ax[1])

diff24 = lst24 - ndvi24
im3 = ax[2].imshow(diff24, cmap=diff_palette)
ax[2].set_title("Difference (Temp - NDVI) 2024")
plt.colorbar(im3, ax=ax[2])

plt.tight_layout()
plt.show()
```

```python
#Histograms with LST
def get_values(array):
    return array.ravel() #non-masked array

fig, ax = plt.subplots(1, 2, figsize=(12, 6))

ax[0].hist(get_values(lst23), bins=50, color="blue", alpha=0.7)
ax[0].set_title("Temperature Histogram 2023")
ax[0].set_xlabel("Temperature")
ax[0].set_ylabel("Frequency")

ax[1].hist(get_values(lst24), bins=50, color="blue", alpha=0.7)
ax[1].set_title("Temperature Histogram 2023")
ax[1].set_xlabel("Temperature")
ax[1].set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
# Scatter plot for 2023 and 2024
plt.figure(figsize=(8, 6))
plt.scatter(get_values(ndvi23), get_values(lst23), alpha=0.5, c="purple")
plt.title("Scatter Plot: NDVI vs Temperature 2023")
plt.xlabel("NDVI")
plt.ylabel("Temperature")
plt.grid(True)
plt.show()

plt.figure(figsize=(8, 6))
plt.scatter(get_values(ndvi24), get_values(lst24), alpha=0.5, c="yellow")
plt.title("Scatter Plot: NDVI vs Temperature 2024")
plt.xlabel("NDVI")
plt.ylabel("Temperature")
plt.grid(True)
plt.show()
```

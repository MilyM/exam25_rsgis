```python
import rasterio
import numpy as np
import matplotlib.pyplot as plt
```

```python
def read_tiff(file_path):
    with rasterio.open(file_path) as src:
        data = src.read(1)  
    return data
```

```python
#path
data_path = r'C:\studia\sem2\pytmat_geoscience\exam\dane\data_kp'
```

```python
#load data
temp1 = read_tiff(f'{data_path}\\t2_lst2023_Jul_Aug.tif')
ndvi1 = read_tiff(f'{data_path}\\t2_ndvi2023_Jul_Aug.tif')

temp2 = read_tiff(f'{data_path}\\t2_lst2024May.tif')
ndvi2 = read_tiff(f'{data_path}\\t2_ndvi2024May.tif')
```

```python
#functions
def display_images(images, titles, cmap='viridis'):
    plt.figure(figsize=(15, 5))
    for i, (img, title) in enumerate(zip(images, titles), 1):
        plt.subplot(1, 3, i)
        plt.imshow(img, cmap=cmap)
        plt.colorbar()
        plt.title(title)
    plt.show()

def plot_histogram(data, title):
    plt.figure()
    plt.hist(data.flatten(), bins=50, color='blue', alpha=0.7)
    plt.title(title)
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.show()

def scatter_plot(x, y, xlabel, ylabel, title):
    plt.figure()
    plt.scatter(x, y, alpha=0.5)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.show()
```

```python
display_images([temp1, ndvi1, temp1 - ndvi1], ['Temperature 2023', 'NDVI 2023', 'Temp - NDVI 2023'])
plot_histogram(temp1, 'Histogram Temperature 2023')
plot_histogram(ndvi1, 'Histogram NDVI 2023')
scatter_plot(ndvi1.flatten(), temp1.flatten(), 'NDVI', 'Temperature', 'Scatter Plot NDVI vs Temperature 2023')
```

```python
display_images([temp2, ndvi2, temp2 - ndvi2], ['Temperature 2024', 'NDVI 2024', 'Temp - NDVI 2024'])
plot_histogram(temp2, 'Histogram Temperature 2024')
plot_histogram(ndvi2, 'Histogram NDVI 2024')
scatter_plot(ndvi2.flatten(), temp2.flatten(), 'NDVI', 'Temperature', 'Scatter Plot NDVI vs Temperature 2024')
```
